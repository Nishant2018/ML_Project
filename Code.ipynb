```python
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.tree import DecisionTreeRegressor
k_values = []
for i in range(50):
    k_values.append(i)

# Initialize a model (e.g., a RandomForestClassifier)
model =RandomForestRegressor()

best_k = None
best_score = 0

for k in k_values:
    if k == 0:
        continue  # Skip k=0
    
    selector = SelectKBest(score_func=f_classif, k=k)
    X_new = selector.fit_transform(x, y)
    
    if X_new.shape[1] == 0:
        continue  # Skip if no features are selected
    
    scores = cross_val_score(model, X_new, y, cv=5)  # 5-fold cross-validation
    mean_score = scores.mean()
    
    print(f"k={k}, Mean Score: {mean_score}")
    
    if mean_score > best_score:
        best_k = k
        best_score = mean_score

print(f"Best k: {best_k}")
```

```python
# Get a boolean mask of selected features
selected_feature_mask = selector.get_support()

# Get the column names of the selected features
selected_feature_names = x.columns[selected_feature_mask]

print("Selected feature names:", selected_feature_names)
```

```python
# Create empty lists to store the column names and p-values
from matplotlib import pyplot as plt
columns = []
p_values = []
import statsmodels.api as sm
# Define a significance level (alpha) for feature selection
alpha = 0.05  # You can adjust this value based on your significance level

# Iterate through each column and perform hypothesis testing
for column in x.columns:
    X_selected = x[column]  # Use only one feature at a time

    # Add a constant to the X matrix (intercept)
    X_selected = sm.add_constant(X_selected)

    # Fit an ordinary least squares (OLS) model
    model = sm.OLS(y, X_selected).fit()

    # Get p-value for the feature
    p_value = model.pvalues[column]

    # Append column name and p-value to the lists
    columns.append(column)
    p_values.append(p_value)

# Create a bar plot to visualize p-values
plt.figure(figsize=(14,24))
bars = plt.barh(columns, p_values, color='skyblue')
plt.xlabel('P-Value')
plt.title('P-Values for Selected Columns')
plt.grid(axis='x', linestyle='--', alpha=0.6)  # Add grid lines

plt.gca().invert_yaxis()  # Invert the y-axis to show the most significant features at the top

# Add p-value labels to the bar lines
for bar, p_value in zip(bars, p_values):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{p_value:.4f}', ha='left', va='center')

# Highlight significant features with a different color
significant_color = 'coral'
threshold = alpha  # Change this if needed
for bar, p_value in zip(bars, p_values):
    if p_value < threshold:
        bar.set_color(significant_color)

plt.show()
```

```python
# Select columns with p-values less than the significance level (e.g., 0.05)
selected_columns = [column for column, p_value in zip(columns, p_values) if p_value < alpha]

print("Selected columns with p-value < 0.05:", selected_columns)
```

```python
not_selected_columns = [column for column, p_value in zip(columns,p_values) if p_value > alpha]

print("Noe selected columns with p-value >0.05 :" , not_selected_columns)
```
